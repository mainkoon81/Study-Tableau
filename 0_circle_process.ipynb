```python
# 1. total (1900)
ed_df = pd.read_csv('./Pre-processing/George_DCC_code/Checkpoints from NACE/Total_Amenity.csv', encoding = "ISO-8859-1")
ed_df.head()
```

```python
ed_df.info()
```

```python
# give point-ID 
ed_df['ID'] = 0

ind = 0
for i in range(len(ed_df['ID'])):
    ind += 1
    ed_df['ID'][i] = ind

# or..ed_df['ID']=np.array(list(np.arange(0, 49))).astype(str)    
    
ed_df['Type'] = 'Point'
ed_df['Degree'] = 0
ed_df['coeff'] = 0
ed_df.tail()
```

## Creating circle df : ed_df2

### 1. Add degree

```python
ed_df2 = ed_df # grow from the original....

ed_df2 = pd.concat([ed_df2]*25) # replicate rows...x 25 (0,15,30,..360)
ed_df2['Type'] = 'Circle'
ed_df2 = ed_df2.sort_index()
```

```python
ed_df2.reset_index(drop=True, inplace=True); ed_df2
```

```python
Degree = np.array(list(np.arange(0,375, 15))*1900)
ed_df2['Degree'] = Degree
```

```python
ed_df2
```

### 2. Add coefficient

```python
ed_df2 = pd.concat([ed_df2]*7) # replicate rows...x 7 (1,2,3,4,5,6,7)
ed_df2 = ed_df2.sort_index()
```

```python
ed_df2.reset_index(drop=True, inplace=True); ed_df2
```

```python
coeff = np.array(list(np.arange(1,8, 1))*47500)
ed_df2['coeff'] = coeff; ed_df2
```

```python
ed_df3 = pd.concat([ed_df, ed_df2])
```
